#  PREPROCESS IMG YOLO

```python
def my_letter_box(img, size=(320, 320)):
    """
    Resize image while preserving aspect rotio, then pad it to taget size.
    The function scales the input image so that it fits within the given `size` (height, width) without distortion. 
    The remain area is filled with a constant padding value (default: 114, 114, 114), commomly used in YOLO preprocessing 
    Args:
        img: (np.array) input data
        size: (tuple [int, int], optional) Target size as (height, width). Default is (320, 320)
    Returns:

    """
    h, w, c = img.shape
    r = min(size[0] / h, size[1] / w)
    new_h, new_w = int(h * r), int(w * r)
    top = int((size[0] - new_h) / 2)
    left = int((size[1] - new_w) / 2)
    bottom = size[0] - new_h - top
    right = size[1] - new_w - left
    img_resize = cv2.resize(img, (new_w, new_h))
    img = cv2.copyMakeBorder(img_resize, top, bottom, left, right, borderType=cv2.BORDER_CONSTANT, value=(114, 114, 114))
    return img, r, left, top

def run_inference_obj(session, input_data):
    """
    Run inference using the ONNX model.
    Args:
        model_path (str): Path to the ONNX model.
        input_data (np.ndarray): Preprocessed input data.
    Returns:
        np.ndarray: Model output.
    """
    _, input_shape = get_input_shape(session)
    input_data, r, left, top = my_letter_box(input_data, (input_shape[2], input_shape[3]))

    input_data = input_data[:, :, ::-1].transpose(2, 0, 1).copy().astype(np.float16)
    input_data = input_data / 255
    input_data = input_data.reshape(1, *input_data.shape)
    # Load the model
    # Get the input name for the model
    input_name = session.get_inputs()[0].name
    # Run inference
    output = session.run(None, {input_name: input_data})
    return output, r, left, top
```
